```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_data_to_load = os.path.join("Resources","schools_complete.csv")
student_data_to_load = os.path.join("Resources","students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Do basic Calculations
total_schools=school_data_complete['school_name'].nunique()
total_students=school_data_complete['student_name'].count()
total_budget=school_data['budget'].sum()
avg_math_score=student_data['math_score'].mean()
avg_score_reading=student_data['reading_score'].mean()
perc_math=(student_data.loc[student_data['math_score']>=70,:].count()['Student ID'])/total_students
perc_read=(student_data.loc[student_data['reading_score']>=70,:].count()['Student ID'])/total_students
perc_both=(student_data.loc[(student_data['reading_score']>=70)&(student_data['math_score']>=70),:].count()['Student ID'])/total_students

#Create a dataframe to hold the above results
district_summary_df=pd.DataFrame({"Total Schools":[total_schools],
                               "Total Students":[total_students],
                               "Total Budget":[total_budget],
                               "Average Math Score":[avg_math_score],
                               "Average Reading Score":[avg_score_reading],
                               "% Passing Math":[perc_math],
                               "% Passing Reading":[perc_read],
                               "% Overall Passing":[perc_both]
                                })

#Clean Format
district_summary_df['Total Students']=district_summary_df['Total Students'].map("{:,.0f}".format)
district_summary_df['Total Budget']=district_summary_df['Total Budget'].map("${:,.2f}".format)
district_summary_df['Average Math Score']=district_summary_df['Average Math Score'].map("{:,.2f}".format)
district_summary_df['Average Reading Score']=district_summary_df['Average Reading Score'].map("{:,.2f}".format)
district_summary_df['% Passing Math']=district_summary_df['% Passing Math'].mul(100).map("{:,.2f}%".format)
district_summary_df['% Passing Reading']=district_summary_df['% Passing Reading'].mul(100).map("{:,.2f}%".format)
district_summary_df['% Overall Passing']=district_summary_df['% Overall Passing'].mul(100).map("{:,.2f}%".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Create dataframe by school

#school name
school_summary_df=school_data_complete.groupby('school_name')

#school by type
school_by_type=school_data[['school_name','type']].set_index('school_name')

#total students
students_by_school=school_summary_df['student_name'].count()

#Total School Budget
budget_by_school=school_summary_df['budget'].mean()

#Budget per student
budget_student=school_summary_df['budget'].mean()/students_by_school

#Average score math
avg_math_score=school_summary_df['math_score'].sum()/school_summary_df['student_name'].count()

#Average Reading score
avg_reading_score=school_summary_df['reading_score'].sum()/students_by_school

#%Passing Math
perc_passing_math_df=school_data_complete.loc[school_data_complete['math_score']>=70,:]
perc_passing_math_df=perc_passing_math_df.groupby('school_name')
perc_passing_math=perc_passing_math_df['math_score'].count()/students_by_school


# %Passing Reading
perc_passing_reading_df=school_data_complete.loc[school_data_complete['reading_score']>=70,:]
perc_passing_reading_df=perc_passing_reading_df.groupby('school_name')
perc_passing_reading=perc_passing_reading_df['reading_score'].count()/students_by_school


#% %Overall Passing
perc_passing_overall_df=school_data_complete.loc[(school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70),:]
perc_passing_overall_df=perc_passing_overall_df.groupby('school_name')
perc_passing_overall=perc_passing_overall_df['reading_score'].count()/students_by_school

# Create dataframe with results
school_summary_df2=pd.DataFrame({
                                 "Total Students":students_by_school,
                                "Total School Budget":budget_by_school,
                                 "Per Student Budget":budget_student,
                                 "Average Math Score":avg_math_score,
                                 "Average Reading Score":avg_reading_score,
                                 "% passing Math":perc_passing_math,
                                  "% passing Reading":perc_passing_reading,
                                "% Overall Passing":perc_passing_overall
 })
school_summary_df2=pd.merge(school_by_type,school_summary_df2,on="school_name",how='outer')
school_summary_df3=school_summary_df2.sort_index()
school_summary_df3=school_summary_df3.rename(columns={'type':'School Type'})


#Format DataFrame numbers
school_summary_df3['Total Students']=school_summary_df2['Total Students'].map("{:,.0f}".format)
school_summary_df3['Total School Budget']=school_summary_df2['Total School Budget'].map("${:,.2f}".format)
school_summary_df3['Per Student Budget']=school_summary_df2['Per Student Budget'].map("${:,.2f}".format)
school_summary_df3['Average Math Score']=school_summary_df2['Average Math Score'].map("{:,.2f}".format)
school_summary_df3['Average Reading Score']=school_summary_df2['Average Reading Score'].map("{:,.2f}".format)
school_summary_df3['% passing Math']=school_summary_df2['% passing Math'].mul(100).map("{:,.2f}%".format)
school_summary_df3['% passing Reading']=school_summary_df2['% passing Reading'].mul(100).map("{:,.2f}%".format)
school_summary_df3['% Overall Passing']=school_summary_df2['% Overall Passing'].mul(100).map("{:,.2f}%".format)
school_summary_df3
```
